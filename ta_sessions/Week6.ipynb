````python
def get_even(numbers):
    # 2 divides exactly an even number
    # when an even number is divided by two, the remainder is 0
    mylist = []
    for number in numbers:
        if number % 2 == 0:
            mylist.append(number)

    return mylist
````

````python
print(get_even([1, 2, 3, 4, 5, 6]))
````

## Normalization

Data normalization is the process of transforming your data by scaling each feature in a given dataset to a particular range. Usually, this range is from 0 to 1. 



Simple feature scaling is basically dividing the value you want to normalize with the maximum value for that particular feature.
Xnew = X/ Xmax

````python
from google.colab import files
uploaded = files.upload()
````

````python
wine_data = pd.read_csv("winequality-red.csv", sep=";")
wine_data.head()
````

````python
# feature by feature
wine_data["fixed acidity"] = wine_data["fixed acidity"] / wine_data["fixed acidity"].max()
wine_data.head()
````

The min-max scaler formula is:

$\bar{x} = \frac{x-x_{min}}{x_{max}- x_{min}}$

With this estimation scale, each feature can be translated separately within the range provided on the training set. The min-max scaler allows you to shrink the data within the range of -1 to 1 when there can be negative values.

````python
# single feature manual

wine_data["fixed acidity"] = (wine_data["fixed acidity"] - wine_data["fixed acidity"].min()) / \
                             (wine_data["fixed acidity"].max() - wine_data["fixed acidity"].min())

wine_data.head()
````

### Z-score Normalization

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

$\text{New value} = \frac{(x\: â€“\: Î¼)}{\sigma}$

where:

x: Original value

$\mu$: Mean of data

$\sigma$: Standard deviation of data

````python
# single feature manual

wine_data["fixed acidity"] = (wine_data["fixed acidity"] - wine_data["fixed acidity"].mean()) / \
                             (wine_data["fixed acidity"].std())
wine_data.head()
````

Reshaping a data from long to wide in python pandas is done with pivot() function. Pivot() function in pandas is one of the efficient function to transform the data from long to wide format.

````python
import numpy as np
#Create a DataFrame
d = {
    'countries':['A','B','C','A','B','C'],
    'metrics':['population_in_million','population_in_million','population_in_million',
                             'gdp_percapita','gdp_percapita','gdp_percapita'],
    'values':[100,200,120,2000,7000,15000]
    }
 
df = pd.DataFrame(d,columns=['countries','metrics','values'])
df
````

Reshape long to wide in pandas python with pivot function:
Pivot function() reshapes the data from long to wide in Pandas python. Countries column is used on index.
Values of Metrics column is used as column names and values of value column is used as its value.

| Name    | Description                                                           | Type                                  | Required / Optional |
|---------|-----------------------------------------------------------------------|---------------------------------------|---------------------|
| index   | Column to use to make new frameâ€™s index. If None, uses existing index | str or object or a list of str        | Optional            |
| columns | Column to use to make new frameâ€™s columns.                            | str or object or a list of str        | Required            |
| values  | Column(s) to use for populating new frameâ€™s values                    | str, object or a list of the previous | Optional            |

````python
# reshape from long to wide in pandas python
 
df2=df.pivot(index='countries', columns='metrics', values='values')
df2
````

To make analysis of data in table easier, we can reshape the data into a more computer-friendly form using Pandas in Python. Pandas.melt() is one of the function to do so..
Pandas.melt() unpivots a DataFrame from wide format to long format.

**melt()** function is useful to massage a DataFrame into a format where one or more columns are identifier variables, while all other columns, considered measured variables, are unpivoted to the row axis, leaving just two non-identifier columns, variable and value.
Syntax :

```
pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None)
```

frame : DataFrame

id_vars[tuple, list, or ndarray, optional] : Column(s) to use as identifier variables.

value_vars[tuple, list, or ndarray, optional]: Column(s) to unpivot. If not specified, uses all columns that are not set as id_vars.

var_name[scalar]: Name to use for the â€˜variableâ€™ column. If None it uses frame.columns.name or â€˜variableâ€™.

value_name[scalar, default â€˜valueâ€™]: Name to use for the â€˜valueâ€™ column.

col_level[int or string, optional]: If columns are a MultiIndex then use this level to melt.

````python
# shape from wide to long with melt function in pandas

df2=pd.melt(df,id_vars=['countries'],var_name='metrics', value_name='values')
df2
````
